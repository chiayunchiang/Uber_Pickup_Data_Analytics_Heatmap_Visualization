# 1. Overview


### 1.1 Introduction
Nowadays, ride-sahring companies, such as Uber, Lyft and Cabify, provide tons of convenience services to the general public. It is easier for us to go everywhere without driving by ourselves. As this kind of services generate huge traffic, we might find some interesting patterns or insights from data.

In this project, I use the dataset- **Uber pickups in New York City** from Kaggle. I first defined a hypothetical business question as the objective, then used explorary data analysis and visualization to answer the question.


### 1.2 Data Description

I use **Uber trip data from 2014** data from **Uber pickups in New York City dataset** as my raw data.   
(Source :https://www.kaggle.com/fivethirtyeight/uber-pickups-in-new-york-city)

This dataset contains Uber trip data from April, 2014 to September, 2014. There are four columns in the dataset as below:   
1. `Date/Time`: The date and time of the Uber pickup
2. `Lat`: The latitude of the Uber pickup
3. `Lon`: The longtitude 
4. `Base`:The TLC base company code affiliated with the Uber pickup



# 2. Objective

As an user of ride-sharing services, sometimes it takes a long time to wait for the driver during off-peak hours because off-peak hours have less drivers than usual time. As a result, our objective is to: **find the potential hot spots during off-peak hours**.

Since we have location and time information of each Uber pickups, we can find hot spots in a specific time period. This gives Uber recommendation on where might have the potenetial customers in a time period. Based on the hot spot recommendation, Uber can allocate their driver to the place near the hot spot, this will shorten the wating time of customer and led to higher customer satification. Besides, the hot spot also give drivers convenience to know which place might have potential customers, which allows drivers to reduce the cost of riding in opposite direction.




# 3. Analysis & Visualization

### 3.1 Import libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap
from folium.plugins import HeatMapWithTime
```

### 3.2 Data pre-processing

```python
# Reads data.
data_name_list = ["apr","may","jun","jul","aug","sep"]
data_file_count = len(data_name_list)
input_file=[]

for i in range(data_file_count):
    input_file.append(pd.read_csv("uber-raw-data-{}14.csv".format(data_name_list[i])))
```

```python
# Merges all the dataframes into one dataframe.
uber_data = pd.concat(input_file)
uber_data.head()
```

```python
# Prints out the data overview.
uber_data.shape
```

```python
# Checks data type
uber_data.dtypes
```

```python
# Checks data information
uber_data.info()
```

```python
# Checks if there is a missing value
uber_data.isnull().sum()
```

```python
# Converts "Data/Time" to datetime type
uber_data["Date/Time"] = pd.to_datetime(uber_data["Date/Time"], 
                                        format="%m/%d/%Y %H:%M:%S")
```

```python
uber_data.dtypes
```

```python
uber_data.head()
```

```python
# Adds new columns of detailed time information by using "Data/Time" columns
uber_data["Month"] = uber_data["Date/Time"].dt.month
uber_data["Day"] = uber_data["Date/Time"].dt.day
uber_data["Hour"] = uber_data["Date/Time"].dt.hour
uber_data["Day_name"] = uber_data["Date/Time"].dt.day_name()
```

```python
uber_data.head()
```

```python
# Sets the order of "Day_name" column in order to benefit following analysis
day_name_order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
uber_data["Day_name"]= pd.Categorical(uber_data["Day_name"], categories=day_name_order, ordered=True)
```

### 3.3 Exploratory Data Analysis

```python
# Defines function which draws bar chart of pickup frequency
def pickup_frequency(df,xlabel):
    df_count = df.pivot_table(index=["{}".format(xlabel)],values="Base",aggfunc="count")
    df_count.plot(kind="bar",figsize=(8,6))
    plt.title("Pickup Frequency by {}".format(xlabel))
    plt.xlabel("{}".format(xlabel))
    plt.ylabel("Frequency")
```

```python
pickup_frequency(uber_data,xlabel="Month")   
```

```python
pickup_frequency(uber_data,xlabel="Day_name") 
```

```python
pickup_frequency(uber_data,xlabel="Hour") 
```

**Brief summary**
1. Between April to September, there are higher pickups on September.
2. Thursday has more pickups,while Sunday has the loweset pickups.
3. There are realtive more pickups between 3am to 9pm. And the peak happened between 4pm to 6pm.
4. There are fewest pickups between 1 am to 5 am.


```python
# Defines function which draw pointplot 
def pointplot(data,x,y,hue):
    fig = plt.figure(figsize=(12,6))
    ax = fig.add_subplot(111)
    sns.pointplot(ax=ax,x=x,y=y,hue=hue, data= data)
    ax.set_title("{} counts by {}".format(hue,x))
    ax.set_xlabel("{}".format(x))
    ax.set_ylabel("{}".format(y))
```

```python
df_month_counts = uber_data.groupby(["Month","Hour"])["Base"]\
.agg("count").to_frame("counts").reset_index().sort_index()
df_month_counts.head()
```

```python
pointplot(df_month_counts,x="Hour",y="counts",hue="Month")
```

Pickups are different between hour period. But there are no specific change caused by month.

```python
df_dayname_counts = uber_data.groupby(["Day_name","Hour"])["Base"].agg("count").to_frame("counts").reset_index().sort_index()
df_dayname_counts.head()
```

```python
pointplot(df_dayname_counts,x="Hour",y="counts",hue="Day_name")
```

**Brief Summary**
1. There is a obvious difference between weekdays and weekends. 
2. The off-peak hours of weekdays are 1am to 3am. 
3. The off-peak hours of weekends are 4am to 6am.

```python
# Checks the pattern of weekdays and weekends whether changed by Month
for i in range(4,10):
    df_dayname_counts_m = uber_data[uber_data["Month"] == i].groupby(["Day_name","Hour"])["Base"]\
    .agg("count").to_frame("counts").reset_index().sort_index()
    pointplot(df_dayname_counts_m,x="Hour",y="counts",hue="Day_name")
```

Although diffenrnt months have different pickups acrossing 24 hours, there still can find the pattern between weekdays and weekends.

**Brief summary**
1. "Month" might not be a factor which influence the pickups changes across hour.
2. Weekdays and weekends have different pickups across hour. It show a obvious pattern between weekdays and weekends.
3. From April to September, the off-peak hours of weekdays appears between 1am to 3am.
4. From April to September, the off-peak hours of weekends appears between 4am to 6am.

According to above summary, we found "Month" is not a sensitivity factor. As a result, we randomly choose one month **"June"** to further analysis to avoid too much pickups points which might lead to unlcearly map. Also, we choose two off-peaks hours points, which are **2 am on weekdays** and **5 am on weekends** as our target period. We will give Uber company the advices of where is the hot spot in these two points.

### 3.4 Data Visualization

```python
# Defines functions which generate New York Map as base map
NY_latitude = 40.7127281
NY_longtitude = -74.0060152

def generate_NY_map(default_location=[NY_latitude,NY_longtitude],default_zoom_start=12):
    NY_map = folium.Map(location = default_location, zoom_start = default_zoom_start)
    return NY_map
```

```python
NY_map = generate_NY_map()
NY_map
```

```python
# Defines function which select specific month
def generate_uber_data_month(data,month):
    return(data[data["Month"]== month])   
```

```python
# Defines function which select specific day_name
def generate_uber_data_dayname(data,dayname):
    return(data[data["Day_name"]== dayname]) 
```

```python
# Defines function which select specific hour
def generate_uber_data_dayname(data,hour):
    return(data[data["Hour"]==hour])
```

```python
# Defines function which generate heatmap
def generate_heatmap(data,month,day_name,hour):
    base_map = generate_NY_map()
    HeatMap(data= data[["Lat","Lon"]].values.tolist(), radius=8).add_to(base_map)
    base_map.add_child(folium.ClickForMarker(popup='Hot Spot(Potential Location)'))
    base_map.save("Uber pickup Heatmap {} {} {}, 2014.html".format(hour,day_name,month))
  
```

We choose **June** for further analysis. According to exploratory data analysis, we select two points, which are **2 am on weekdays** and **5 am on weekends** as our targets to find hot points. 


```python
weekends = ["Saturday","Sunday"]
uber_data_jun = generate_uber_data_month(uber_data,6)
uber_data_jun_weekday = uber_data_jun[~uber_data_jun["Day_name"].isin(weekends)]
uber_data_jun_weekday_2 = generate_uber_data_dayname(uber_data_jun_weekday ,2)
```

```python
# Generates heatmap at 2am on June on Weekdays,2014
generate_heatmap(uber_data_jun_weekday_7,month="on June",day_name="on Weekdays",hour="at 2am")
```

```python
uber_data_jun_weekends = uber_data_jun[uber_data_jun["Day_name"].isin(weekends)]
uber_data_jun_weekends_5 = generate_uber_data_dayname(uber_data_jun_weekday ,5)
```

```python
# Generates heatmap at 5am on June on Weekends,2014
generate_heatmap(uber_data_jun_weekends_0,month="on June",day_name="on Weekends",hour="at 5am")
```

**Brief summary**   
  
Heatmap results are shown on **html**. at our default zoom (12), we identify the potential hot spot on the map.    
1. We found at 2am on weekdays, there are five hot spots in Manhattan: **Upper East Side**, **Upper West Side**, **East Village**, **West Village** and **Lower Manhattan**.
2. We found at 5am on weekends, there are two hot spots in Manhattan: **West Village** and **Hell's Kitchen**.

This could be a recommendation for Uber company to allocate to their drivers.


Besides, We visualize the heatmap with time. Therefore, Uber comapny can see how hot spot changes over by the time.

```python
# Define function which plot heatmap with time changes
# add below code to the output html 
# <script src="https://cdn.jsdelivr.net/npm/leaflet-timedimension@1.1.0/dist/leaflet.timedimension.min.js"></script>
# Reference: https://gis.stackexchange.com/questions/341004/folium-plugins-heatmapwithtime-no-longer-produces-an-output
def generate_heatmap_with_time(data,month,day_name):
    
    base_map = generate_NY_map()
    data_with_time = []
    
    for hour in data["Hour"].unique():
        hour_list = []
        point = data[data["Hour"]==hour]
        for lat,lon in zip(point["Lat"],point["Lon"]):
            location=[lat,lon]
            hour_list.append(location)
            
        data_with_time.append(hour_list)
    
    HeatMapWithTime(data = data_with_time,auto_play = True, radius=6).add_to(base_map)
    base_map.save("Uber pickup Heatmap with time {} {},2014.html".format(day_name,month))
```

```python
generate_heatmap_with_time(uber_data_jun_weekday,"on June","on Weekdays")
```

```python
generate_heatmap_with_time(uber_data_jun_weekends,"on June","on Weekends")
```

We can easily see how pickups points changes by the time in our ""heatmap_with_time"".

# 4. Discussion

In this project, we find out pickups hot spot by using visualization map, and mark the hot spot manually. However, here comes 3 prolems as following:   
1. Since the density of heatmap is based on default zoom value, the hot spot will change if we change the defautl zoom value.   

2. Since we mark the hot spot manually, we can not make sure whtether selected point is really a hot spot if there are tons of pickups piles in the map (which results in difficulity to identify the right point).  

3. It might be more useful if we can give Uber company a "Hot spot Area" recommendation, so that they can identify potential area for Uber driver to stay or waiting for customers.
